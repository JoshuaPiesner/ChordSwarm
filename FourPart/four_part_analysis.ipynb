```python
# In a Jupyter‐style cell (so VS Code’s Interactive Window or .ipynb), first enable the widget backend:
%matplotlib widget

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from ipywidgets import FloatSlider, VBox, Output
from IPython.display import display

# Prepare an Output widget that will hold our figure
plot_out = Output()

# Dummy data (swap in your df)
import pandas as pd, numpy as np

# The routine that (re-)draws the 3D scatter into plot_out
def redraw(elev, azim):
    with plot_out:
        plot_out.clear_output(wait=True)
        fig = plt.figure(figsize=(8,6))
        ax  = fig.add_subplot(111, projection='3d')
        sc  = ax.scatter(
            df['chord_sum_mean'],
            df['continu_sum_mean'],
            df['chord_strength'],
            c=df['continu_strength'],
            cmap='plasma',
            alpha=0.7
        )
        ax.set_xlabel('Chord Sum Mean')
        ax.set_ylabel('Continu Sum Mean')
        ax.set_zlabel('Chord Strength')
        ax.set_title('3D Scatter')
        ax.view_init(elev=elev, azim=azim)
        fig.colorbar(sc, label='Continu Strength')
        plt.show()

# Sliders, initialized to a sensible default
elev_slider = FloatSlider(min=-180, max=180, step=1, value=30, description='Elevation')
azim_slider = FloatSlider(min=-180, max=180, step=1, value=-60, description='Azimuth')

# Wire up callbacks to re-plot on change
def on_change(change):
    redraw(elev_slider.value, azim_slider.value)

elev_slider.observe(on_change, names='value')
azim_slider.observe(on_change, names='value')

# Do an initial draw
redraw(elev_slider.value, azim_slider.value)

# Display sliders and plot side by side (or stacked)
controls = VBox([elev_slider, azim_slider])
display(controls, plot_out)
```

```python
# 1) Enable the interactive widget backend
%matplotlib widget

# 2) Imports
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D             # registers the 3D projection
from ipywidgets import FloatSlider, VBox, Output
from IPython.display import display
import pandas as pd
import numpy as np


# 4) Prepare the Output widget for the plot
plot_out = Output()

# 5) The redraw function (re-creates the figure)
def redraw(elev, azim):
    with plot_out:
        plot_out.clear_output(wait=True)
        fig = plt.figure(figsize=(8,6))
        ax  = fig.add_subplot(111, projection='3d')
        
        # Z-axis is continu_strength, color is chord_strength
        sc = ax.scatter(
            df['chord_sum_mean'],     # X
            df['continu_sum_mean'],   # Y
            df['continu_strength'],   # Z
            c=df['chord_strength'],   # color
            cmap='plasma',
            alpha=0.6,
        )
        
        # Labels & title
        ax.set_xlabel('Chord Sum Mean')
        ax.set_ylabel('Continu Sum Mean')
        ax.set_zlabel('Continu Strength')
        ax.set_title('3D: Chord Sum vs Continu Sum vs Continu Strength')
        
        # Apply the current slider angles
        ax.view_init(elev=elev, azim=azim)
        
        # Colorbar now shows chord_strength
        fig.colorbar(sc, label='Chord Strength')
        plt.show()

# 6) Create the sliders
elev_slider = FloatSlider(
    value=30, min=-180, max=180, step=1,
    description='Elevation'
)
azim_slider = FloatSlider(
    value=-60, min=-180, max=180, step=1,
    description='Azimuth'
)

# 7) Wire up the callbacks
def on_change(_):
    redraw(elev_slider.value, azim_slider.value)

elev_slider.observe(on_change, names='value')
azim_slider.observe(on_change, names='value')

# 8) Initial draw + display
redraw(elev_slider.value, azim_slider.value)
display(VBox([elev_slider, azim_slider]), plot_out)
```
